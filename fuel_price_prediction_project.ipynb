# Fuel Price Prediction — EDA + Machine Learning Portfolio Project

This notebook uses two datasets:

- `price_history.csv`: historical fuel prices by station/node and fuel type
- `stations.csv`: station metadata such as brand, location, supermarket/motorway flags, and opening details

## Project goal
Build a complete data science project that:

1. Loads and understands the raw data
2. Cleans and joins station + price data
3. Performs exploratory data analysis (EDA)
4. Finds business insights about fuel pricing
5. Builds baseline and machine learning models to predict fuel prices
6. Explains limitations and next steps in a portfolio-friendly way

## Target variable
`price_pence` — fuel price in pence per litre.

## 1. Import libraries

These are common Python libraries used in data analysis and machine learning projects.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

pd.set_option('display.max_columns', 100)
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')
```

## 2. Load the data

Keep the original files unchanged. We will create cleaned versions in memory.

```python
price_path = 'price_history.csv'
stations_path = 'stations.csv'

prices = pd.read_csv(price_path, parse_dates=['recorded_at', 'source_updated_at'])
stations = pd.read_csv(stations_path, parse_dates=['first_seen', 'last_seen', 'updated_at'])

print('Price history shape:', prices.shape)
print('Stations shape:', stations.shape)

display(prices.head())
display(stations.head())
```

## 3. Initial data checks

Before modelling, check:

- column types
- missing values
- duplicate rows
- date ranges
- unique stations and fuel types

```python
print('--- Price data info ---')
display(prices.info())

print('--- Station data info ---')
display(stations.info())

print('
Missing values in price data:')
display(prices.isna().sum().sort_values(ascending=False))

print('
Missing values in station data:')
display(stations.isna().sum().sort_values(ascending=False).head(15))

print('
Date range:')
print(prices['recorded_at'].min(), 'to', prices['recorded_at'].max())

print('
Fuel types:')
display(prices['fuel_type'].value_counts())

print('
Unique stations in price history:', prices['node_id'].nunique())
print('Unique stations in stations table:', stations['node_id'].nunique())
```

## 4. Join price data with station metadata

`node_id` appears in both tables, so we use it to enrich each price record with station details.

```python
# Standardise text fields before joining / modelling
stations_clean = stations.copy()
stations_clean['brand_name_clean'] = stations_clean['brand_name'].fillna('Unknown').str.strip().str.title()
stations_clean['city_clean'] = stations_clean['city'].fillna('Unknown').str.strip().str.title()
stations_clean['county_clean'] = stations_clean['county'].fillna('Unknown').str.strip().str.title()

station_cols = [
    'node_id', 'brand_name_clean', 'city_clean', 'county_clean',
    'latitude', 'longitude', 'is_motorway', 'is_supermarket',
    'is_temporarily_closed', 'is_permanently_closed'
]

df = prices.merge(stations_clean[station_cols], on='node_id', how='left')

print('Joined shape:', df.shape)
print('Rows without station metadata:', df['brand_name_clean'].isna().sum())
display(df.head())
```

## 5. Data quality and outlier check

The raw price data contains implausible values such as 0, 1, and very high prices like 1999 pence/litre. For this project, we will use a simple business rule and keep prices between **100 and 300 pence per litre**.

This is a modelling decision. In a real project, you would confirm the correct range with a domain expert or source documentation.

```python
print('Raw price summary:')
display(df['price_pence'].describe())

plt.figure(figsize=(8, 4))
df['price_pence'].hist(bins=100)
plt.title('Raw price distribution')
plt.xlabel('Price, pence per litre')
plt.ylabel('Record count')
plt.show()

raw_rows = len(df)
df = df[(df['price_pence'] >= 100) & (df['price_pence'] <= 300)].copy()
removed_rows = raw_rows - len(df)

print(f'Removed {removed_rows:,} implausible rows')
print('Cleaned shape:', df.shape)

display(df['price_pence'].describe())
```

## 6. Feature engineering

Create date-based features that a model can use:

- day index from the first date
- day of week
- month
- weekend flag

```python
df['date'] = df['recorded_at'].dt.date
df['day_index'] = (df['recorded_at'].dt.normalize() - df['recorded_at'].dt.normalize().min()).dt.days
df['day_of_week'] = df['recorded_at'].dt.dayofweek
df['month'] = df['recorded_at'].dt.month
df['is_weekend'] = df['day_of_week'].isin([5, 6])

# Useful for charts
df['fuel_type'] = df['fuel_type'].astype(str)

display(df[['recorded_at', 'date', 'day_index', 'day_of_week', 'month', 'is_weekend']].head())
```

# Exploratory Data Analysis

This section helps answer: *what drives fuel prices in this dataset?*

## 7. Fuel type analysis

Different fuel types have different average prices. This is likely one of the strongest predictors.

```python
fuel_summary = (
    df.groupby('fuel_type')['price_pence']
      .agg(records='count', mean_price='mean', median_price='median', min_price='min', max_price='max')
      .sort_values('records', ascending=False)
)

display(fuel_summary)

fuel_summary['mean_price'].sort_values().plot(kind='barh', figsize=(8, 4))
plt.title('Average price by fuel type')
plt.xlabel('Average price, pence per litre')
plt.ylabel('Fuel type')
plt.show()
```

## 8. Daily price trend

A time trend helps us understand whether prices are rising or falling across the dataset period.

```python
daily_fuel = (
    df.groupby(['date', 'fuel_type'])['price_pence']
      .mean()
      .reset_index()
)

plt.figure(figsize=(12, 5))
for fuel in daily_fuel['fuel_type'].unique():
    temp = daily_fuel[daily_fuel['fuel_type'] == fuel]
    plt.plot(temp['date'], temp['price_pence'], label=fuel)

plt.title('Daily average fuel price by fuel type')
plt.xlabel('Date')
plt.ylabel('Average price, pence per litre')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 9. Supermarket vs non-supermarket stations

This is useful business insight because supermarket fuel stations are often perceived as cheaper.

```python
supermarket_summary = (
    df.groupby('is_supermarket')['price_pence']
      .agg(records='count', mean_price='mean', median_price='median')
)

display(supermarket_summary)

by_fuel_supermarket = (
    df.groupby(['fuel_type', 'is_supermarket'])['price_pence']
      .mean()
      .unstack()
)

display(by_fuel_supermarket)

by_fuel_supermarket.plot(kind='bar', figsize=(9, 5))
plt.title('Average price by fuel type: supermarket vs non-supermarket')
plt.xlabel('Fuel type')
plt.ylabel('Average price, pence per litre')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 10. Motorway vs non-motorway stations

Motorway stations are usually more expensive. This feature is likely predictive.

```python
motorway_summary = (
    df.groupby('is_motorway')['price_pence']
      .agg(records='count', mean_price='mean', median_price='median')
)

display(motorway_summary)

motorway_summary['mean_price'].plot(kind='bar', figsize=(6, 4))
plt.title('Average price: motorway vs non-motorway')
plt.xlabel('Is motorway station?')
plt.ylabel('Average price, pence per litre')
plt.xticks(rotation=0)
plt.show()
```

## 11. Brand analysis

Brand can capture pricing strategy, station type, and customer segment. To avoid misleading results, only compare brands with enough records.

```python
brand_summary = (
    df.groupby('brand_name_clean')
      .agg(records=('price_pence', 'size'), mean_price=('price_pence', 'mean'))
      .query('records >= 2000')
      .sort_values('mean_price')
)

display(brand_summary.head(20))
display(brand_summary.tail(20))

brand_summary['mean_price'].head(15).plot(kind='barh', figsize=(8, 5))
plt.title('Lowest average price brands, minimum 2,000 records')
plt.xlabel('Average price, pence per litre')
plt.ylabel('Brand')
plt.show()
```

## 12. Location analysis

Latitude and longitude can help the model learn regional pricing differences. A simple scatter plot gives a quick visual check.

```python
location_sample = df.sample(min(25000, len(df)), random_state=42)

plt.figure(figsize=(7, 8))
plt.scatter(location_sample['longitude'], location_sample['latitude'], s=3, alpha=0.25)
plt.title('Station price records by location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

# Predictive Modelling

We will build this like a professional ML project:

1. Use a **time-based split** instead of a random split, because future prices should be predicted from past data.
2. Build a simple **baseline model**.
3. Build a machine learning model using numeric and categorical features.
4. Compare metrics.

## Important note about prediction target
This model predicts the price for a station/fuel/date combination based on historical patterns. It does **not** know future crude oil prices, taxes, wholesale costs, or live market shocks unless those features are added later.

## 13. Prepare modelling data

To keep training fast and beginner-friendly, we reduce high-cardinality categories by keeping the most common values and grouping the rest as `Other`.

```python
model_df = df.dropna(subset=['brand_name_clean', 'latitude', 'longitude']).copy()

# Reduce high-cardinality categories for a beginner-friendly first model
for col, top_n in [('brand_name_clean', 25), ('city_clean', 40), ('county_clean', 40)]:
    top_values = model_df[col].value_counts().head(top_n).index
    model_df[col] = model_df[col].where(model_df[col].isin(top_values), 'Other')

categorical_features = [
    'fuel_type', 'brand_name_clean', 'city_clean', 'county_clean',
    'is_motorway', 'is_supermarket', 'is_weekend'
]

numeric_features = [
    'latitude', 'longitude', 'day_index', 'day_of_week', 'month'
]

target = 'price_pence'

print('Modelling rows:', model_df.shape[0])
print('Features:', categorical_features + numeric_features)
```

## 14. Time-based train/test split

The test set is the most recent 14 days in the dataset. This is more realistic than a random split.

```python
cutoff_date = model_df['recorded_at'].max() - pd.Timedelta(days=14)

train_df = model_df[model_df['recorded_at'] < cutoff_date].copy()
test_df = model_df[model_df['recorded_at'] >= cutoff_date].copy()

print('Cutoff date:', cutoff_date)
print('Train rows:', train_df.shape[0])
print('Test rows:', test_df.shape[0])

X_train = train_df[categorical_features + numeric_features]
y_train = train_df[target]

X_test = test_df[categorical_features + numeric_features]
y_test = test_df[target]
```

## 15. Baseline model

A baseline is essential in portfolio projects. Here the baseline predicts the average historical price for each fuel type.

```python
fuel_mean = train_df.groupby('fuel_type')[target].mean()
global_mean = y_train.mean()

baseline_pred = X_test['fuel_type'].map(fuel_mean).fillna(global_mean)

def evaluate_model(name, y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred) ** 0.5
    r2 = r2_score(y_true, y_pred)
    return {'model': name, 'MAE': mae, 'RMSE': rmse, 'R2': r2}

results = []
results.append(evaluate_model('Baseline: fuel-type mean', y_test, baseline_pred))

pd.DataFrame(results)
```

## 16. Machine learning model: Ridge Regression

Ridge Regression is a strong first model for tabular data with many one-hot encoded categorical features. It is easy to explain and trains quickly.

```python
preprocess = ColumnTransformer(
    transformers=[
        ('categorical', OneHotEncoder(handle_unknown='ignore'), categorical_features),
        ('numeric', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), numeric_features)
    ]
)

ridge_model = Pipeline(steps=[
    ('preprocess', preprocess),
    ('model', Ridge(alpha=2.0, solver='lsqr'))
])

ridge_model.fit(X_train, y_train)
ridge_pred = ridge_model.predict(X_test)

results.append(evaluate_model('Ridge Regression', y_test, ridge_pred))
results_df = pd.DataFrame(results).sort_values('MAE')
display(results_df)
```

## 17. Actual vs predicted prices

This plot shows whether predictions are generally close to real prices.

```python
plot_df = pd.DataFrame({
    'actual': y_test.values,
    'predicted': ridge_pred
}).sample(min(10000, len(y_test)), random_state=42)

plt.figure(figsize=(6, 6))
plt.scatter(plot_df['actual'], plot_df['predicted'], s=5, alpha=0.25)
plt.plot([plot_df['actual'].min(), plot_df['actual'].max()],
         [plot_df['actual'].min(), plot_df['actual'].max()])
plt.title('Actual vs predicted fuel prices')
plt.xlabel('Actual price, pence per litre')
plt.ylabel('Predicted price, pence per litre')
plt.show()
```

## 18. Error analysis

Look at where the model performs badly. This helps you improve the next version.

```python
error_df = test_df.copy()
error_df['prediction'] = ridge_pred
error_df['absolute_error'] = (error_df[target] - error_df['prediction']).abs()

print('Largest prediction errors:')
display(error_df[[
    'recorded_at', 'fuel_type', 'brand_name_clean', 'city_clean', 'is_motorway',
    'is_supermarket', 'price_pence', 'prediction', 'absolute_error'
]].sort_values('absolute_error', ascending=False).head(20))

error_by_fuel = error_df.groupby('fuel_type')['absolute_error'].agg(['count', 'mean', 'median']).sort_values('mean', ascending=False)
display(error_by_fuel)
```

## 19. Make a single prediction

This example shows how you could later turn the model into a small app or API.

```python
example = X_test.sample(1, random_state=7)
predicted_price = ridge_model.predict(example)[0]

print('Input example:')
display(example)
print(f'Predicted price: {predicted_price:.2f} pence per litre')
```

- Premium fuels are generally more expensive than standard fuels.
- Motorway stations are significantly more expensive on average.

The first ML model uses:

- fuel type
- brand
- city/county
- latitude/longitude

- crude oil prices
- wholesale fuel costs
- taxes/duties
- competitor prices nearby
- traffic level
- live supply shocks
- station-specific pricing strategy over longer time periods
